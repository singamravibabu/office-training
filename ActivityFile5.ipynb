# The 'if' statement (construct)

- The `if` statement is a conditional branching statement.

##### Syntax:
```
if <bool_exp>:
    block
[elif <bool_exp>:
    block]...
[else:
    block]

```
if (required)		1
elif (optional)		0 to n
else (optional)		0 or 1


block	
    - python statements
	- indented statements
	- all the statements in the block should be equally indented
```

````python
# Code with only 'if' clause
````

````python
age = 17

if age >= 18:
    print("You are eligible to vote.")
    print("Voting is on November 30th.")
    print("See you at the polling station!")
print("Bye!!!")
````

````python
age = 35

if age >= 18:
    print("You are eligible to vote.")
    print("Voting is on November 30th.")
    print("See you at the polling station!")
else:
    print("You are not eligible to vote yet.")
    print(f"Wait for {18-age} more years.")
````

````python
marks = 104
if marks >= 90 and marks <= 100:
    print("Grade A")
elif marks >= 80 and marks < 90:
    print("Grade B")
elif marks >= 70 and marks < 80:
    print("Grade C")
elif marks >= 60 and marks < 70:
    print("Grade D")
elif marks >= 50 and marks < 60:
    print("Grade E")
elif marks >= 0 and marks < 50:
    print("Fail")
else:
    print("Not a valid entry!!")
````

````python
marks = 104
if 90 <= marks <= 100:
    print("Grade A")
elif 80 <= marks < 90:
    print("Grade B")
elif 70 <= marks < 80:
    print("Grade C")
elif 60 <= marks < 70:
    print("Grade D")
elif 50 <=  marks < 60:
    print("Grade E")
elif 0 <= marks < 50:
    print("Fail")
else:
    print("Not a valid entry!!")
````

````python
marks = 99
if marks >= 80:
    if marks >= 90:
        print("Grade A")
    else:
        print("Grade B")
elif marks >= 60:
    if marks >= 70:
        print("Grade C")
    else:
        print("Grade D")
elif marks >= 50:
    print("Grade E")
else:
    print("Fail")
````

````python
city = input("Enter city name: ")
match city:
    case "New York":
        print("The Big Apple")
    case "Los Angeles":
        print("City of Angels")
    case "Chicago":
        print("The Windy City")
    case "Houston":
        print("Space City")
    case _:
        print("Unknown City")   
````

````python
city = input("Enter city name: ")
cities = {
    "New York": "The Big Apple",
    "Los Angeles": "City of Angels",
    "Chicago": "The Windy City",
    "Houston": "Space City"
    }
print(cities.get(city, "Unknown City"))
````
