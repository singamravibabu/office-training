# WORKING WITH TUPLES

- Like lists, tuples are data structures taht can contain multiple items.
- Tuple is an immutable object.

- Syntax for creating a tuple
    - (item1, item2, item3, item4, ...)
    - my_tuple = (item1, item2, item3, item4, ...)

```python
# tuple of departments
dept = ('accounts', 'finance', 'sales', 'marketing', 'hr', 'it', 'admin')
```

```python
# Accessing items in a tuple just like items in a list using indexing
print(dept[0])  # Output: accounts
print(dept[3])  # Output: marketing
print(dept[-1]) # Output: admin
# Tuples are immutable, so the following line would raise an error if uncommented
# dept[0] = 'accounting'  # This will raise a TypeError
```

```python
d1, d2, d3, d4, d5, d6, d7 = dept
```

```python
print(d1, d4, d7)
```

```python
# because the tuples are immutable, they are efficient in terms of memory and performance compared to lists.
```
